```python
def coefficient_of_corelation(height, weight):
  result = 0
  E_Weight = 0
  E_Height = 0
  D_Weight = 0
  D_Height = 0
  E_HW = 0
  for i in range(height.size):
    E_HW += (1 / height.size) * height[i] * weight[i]
    E_Height += (1 / height.size) * height[i]
    E_Weight += (1 / weight.size) * weight[i]
  for i in range(height.size):
    D_Height += (1 / height.size) * (height[i] - E_Height) ** 2
    D_Weight += (1 / weight.size) * (weight[i] - E_Weight) ** 2
  result = (E_HW - E_Height * E_Weight) / (math.sqrt(D_Height) * math.sqrt(D_Weight))
  return result

height = np.array([170, 182,  183,  180, 175, 181, 187, 181, 178, 187])
weight = np.array([66, 74, 77,  72, 67, 77, 76, 77, 72, 76])
print("%.2f" % coefficient_of_corelation(height, weight))
```

```python
def mean_and_dispersion(array, order):
  E = 0
  D = 0
  n = len(array)
  for i in range(n):
    E += (1 / n) * array[i] ** order
  for i in range(n):
    D += (1 / n) * (array[i] - E) ** 2
  return (E, D)

def get_variation_row(array):
  n = len(array)
  k = 0
  variation = [0 for i in range(n)]
  count = np.array([0 for i in range(max(array) + 1)])
  for i in range(n):
    count[array[i]] += 1 
  for i in range(count.size):
    for j in range(count[i]):
      variation[k] = i
      k += 1
  #print(variation)
  return variation

def median(array):
  result = 0
  variation = get_variation_row(array)
  if array.size % 2 == 0:
    result += variation[int(array.size / 2)] + variation[int(array.size / 2) - 1]
  else:
    result += variation[int(array.size / 2)]
  return result

def excess(array):
  return mean_and_dispersion(array, 3)[0] / (mean_and_dispersion(array, 3)[1] ** 1.5)  
def asymmetry(array):
  return ((mean_and_dispersion(array, 4)[0] / (mean_and_dispersion(array, 4)[1] ** 2)) - 3)
#mean_and_dispersion(np.array([1,4,7,8,-1,0]))[1]
```
